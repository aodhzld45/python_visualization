```python
# 상관 분석 - 두 변수의 관계 분석하기
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 

economics = pd.read_csv('economics.csv')
print(economics)

# economics의 상관 행렬 만들기 1. unemploy - 실업자 수  / pce - 개인 소비 지출
correlation_matrix = economics[['unemploy', 'pce']].corr()

# 유의 확률 구하기
# 1. 상관 분석
stats.pearsonr(economics['unemploy'], economics['pce'])

# 상관행렬 히트맵 만들기
mtcars = pd.read_csv('mtcars.csv')
car_cor = mtcars.corr() # 상관행렬 만들기
car_cor = round(car_cor, 2)
car_cor

# 2. 히트맵 만들기 
plt.rcParams.update({'figure.dpi' : '120', # 해상도 설정
                     'figure.figsize' : [7.5, 5.5]}) # 가로 세로 크기 설정

# sns.heatmap(car_cor,
#            annot = True, # 상관계수 표시
#            cmap = 'RdBu') # 컬러맵

# 3. mask 만들기 -> 표시하고 싶은 곳에 0으로 표현 / 지우고 싶은 곳에는 1로 표현
mask = np.zeros_like(car_cor)

# 3-1 오른쪽 위 대각 행렬을 1로 바꾸기
mask[np.triu_indices_from(mask)] = 1 
mask

# 4. 히트맵에 mask 적용하기
# sns.heatmap(data = car_cor,
#            annot = True, # 상관계수 표시
#            cmap = 'RdBu',
#            mask = mask) # 컬러맵

# 5. 빈 행과 열 제거하기 
mask_new = mask[1:, :-1] # mask 첫 번쨰행, 마지막 열 제거
cor_new = car_cor.iloc[1:, :-1] # 상관행렬 첫 번쨰 행, 마지막 열 제거

sns.heatmap(data = cor_new, # 데이터
           annot = True, # 상관계수 표시
           cmap = 'RdBu', # 컬러맵
           mask = mask_new, # mask 적용
           linewidths = .5, # 경계구분선 추가
           vmax = 1,        # 가장 진한 파란색으로 표현할 최댓값
           vmin = -1,       # 가장 진한 빨간색으로 표현할 최솟값
           cbar_kws = {'shrink': .5}) # 범례크기 줄이기
```
