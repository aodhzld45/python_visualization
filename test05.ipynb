```python
# 115p 혼자서 해보기 - mpg 데이터를 이용한 분석 문제
import pandas as pd

mpg = pd.read_csv('mpg.csv') # mpg.csv 파일 불러오기
mpg_copy = mpg.copy() # mpg_copy에 mpg.csv 복사 

# mpg_copy.rename(columns = {'cty':'city','hwy':'highway'}) -> 원복 객체에 영향을 끼치지 않음
mpg_copy = mpg_copy.rename(columns = {'cty':'city','hwy':'highway'}) # 새로운 객체를 생성하여 다시 return을 해줘야하는 rename 메서드의 특성

mpg_copy['total'] = (mpg_copy['city'] + mpg_copy['highway']) / 2

# 조건물을 활용하여 파생변수 만들기

# 1. 기준값 정하기
print(mpg_copy['total'].describe()) # 요약 통계량 출력
mpg_copy['total'].plot.hist()

import numpy as np # 조건문을 사용하기 위한 numpy 패키지 import
mpg_copy['test'] = np.where(mpg_copy['total'] >= 20, 'pass', 'fail') # total 값 옆에 test 위치 -> 20 이상이면 pass 그렇지 않으면 fail 부여
counts = mpg_copy['test'].value_counts() # test columns 빈도 수 구하기
type(counts)
# counts.plot.bar(rot = 0)

# 중첩 조건문 활용하기 p = 123
# 30 이상이면 A / 20 ~ 29 B / 20 미만 C
mpg_copy['grade'] = np.where(mpg_copy['total'] >= 30, 'A',
                    np.where(mpg_copy['total'] >= 20, 'B', 'C'))
    
mpg_copy

mpg_copy.tail(20)

grade_counts = mpg_copy['grade'].value_counts().sort_index() # grade columns 빈도 수 구하기
print(grade_counts)
grade_counts.plot.bar(rot = 0) # 
grade_counts.plot.pie()

```
